```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads.head(10)
```

```python
recent_grads.tail(5)
```

```python
recent_grads.describe(include='all')
```

```python
# look up number of rows
raw_data_count = recent_grads.shape[0]
raw_data_count
```

```python
# drop rows with missing values
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
cleaned_data_count
```

```python
recent_grads.plot(x='Median', y='Sample_size', kind='scatter', title='Median Earnings', figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment Rate for Recent Grads', figsize=(5,10))
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median Earnings for full-time employment', figsize=(5,10))
ax.set_ylabel('Median Earnings')
```

```python
# Do female grads experience greater unemployment?
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter')
```

```python
# show distribution of median income. What is the most common income range for recent grads?
recent_grads['Median'].hist(bins=25, range=(0,110000))
```

```python
# distribution of Men
recent_grads['Men'].hist(bins=25, range=(0,20000))
```

```python
# distribution of Women
recent_grads['Women'].hist(bins=25, range=(0,20000))
```

```python
from pandas.plotting import scatter_matrix
```

```python
#
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

```python
# Compare share of women in engineering majors vs others
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

```python
# Compare unemployment rate by major
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[163:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

```python
# distribution of median salaries
fig2, ax2 = plt.subplots()
ax2.set_title('Distribution of Median Income')
ax2.boxplot(recent_grads['Median'])
```

```python
# distribution of unemployment
fig3, ax3 = plt.subplots()
ax3.set_title('Distribution of Unemployment')
ax3.boxplot(recent_grads['Unemployment_rate'])
```
